```python
df_incidentes= pd.read_csv('Police_Department_Incidents.csv')
df_incidentes.head()
```

```python
df_incidentes.shape
```

```python
limite = 100
df_incidentes = df_incidentes.iloc[0:limite, : ]
```

```python
df_incidentes.shape
```

```python
latitude = 37.77
longitude = -122.42
sf_map = folium.Map(Location=[latitude, longitude], zoom_start=10)
sf_map
```

```python
incidentes = folium.map.FeatureGroup()
for lat, lng in zip (df_incidentes.Y, df_incidentes.X):
    incidentes.add_child(
    folium.CircleMarker(
    [lat,lng],
    radius =5,
    color='yellow',
    fill =True,
    fill_color ='blue',
    fill_opacity = 0.6)
    )
        
sf_map.add_child(incidentes)
```

```python
incidentes = folium.map.FeatureGroup()
for lat, lng in zip (df_incidentes.Y, df_incidentes.X):
    incidentes.add_child(
    folium.CircleMarker(
    [lat,lng],
    radius =5,
    color='yellow',
    fill =True,
    fill_color ='blue',
    fill_opacity = 0.6)
    )

latitudes = list(df_incidentes.Y)
longitudes = list(df_incidentes.X)
labels = list(df_incidentes.Category)

for lat, lng, label in zip(latitudes, longitudes, labels):
    folium.Marker([lat,lng], popup=label).add_to(sf_map)
        
sf_map.add_child(incidentes)
```

```python
latitude = 37.77
longitude = -122.42
sf_map = folium.Map(Location=[latitude, longitude], zoom_start=10)
for lat, lng, label in zip (df_incidentes.Y, df_incidentes.X, df_incidentes.Category):
    folium.CircleMarker(
    [lat,lng],
    radius =5,
    color='yellow',
    fill =True,
    fill_color ='blue',
    fill_opacity = 0.6,
    popup = label).add_to(sf_map)
sf_map
```

```python
from folium import plugins
latitude = 37.77
longitude = -122.42
sf_map = folium.Map(Location=[latitude, longitude], zoom_start=12)
incidentes = plugins.MarkerCluster().add_to(sf_map)

for lat, lng, label, in zip (df_incidentes.Y, df_incidentes.X, df_incidentes.Category):
    folium.Marker(
    location=[lat, lng],
    icon= None,
    popup=label).add_to(incidentes)
    
sf_map
```

## Mapas Choropleth

```python
df_can= pd.read_excel('Canada.xlsx', sheet_name='Canada by Citizenship',
                     skiprows=range(20),
                     skipfooter=2)
df_can.head()
```

```python
df_can.shape
```

```python
df_can.drop(['AREA','REG','DEV','Type','Coverage'], axis=1, inplace= True)
```

```python
df_can.rename(columns={'OdName':'Country', 'AreaName':'Continent','RegName':'Region'},
             inplace=True)
```

```python
df_can.columns= list (map(str, df_can.columns))
```

```python
df_can['Total']=df_can.sum(axis=1)
```

```python
df_can.head()
```

```python
world_geo = r'world_countries.json'
```

```python
world_map=folium.Map(location=[0,0], zoom_start=2, tiles='Mapbox Bright')
world_map.choropleth(
geo_data= world_geo,
data= df_can,
columns = ['Country','Total'],
key_on='feature.properties.name',
fill_color='YlOrRd',
fill_opacity = 0.7,
line_opacity=0.2,
legend_name='Inmigracion de los paises a canada')
world_map
```
